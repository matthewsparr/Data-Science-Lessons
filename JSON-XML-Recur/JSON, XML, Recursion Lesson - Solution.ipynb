## Recursion

Write a function to check if a given string is a palindrome.

```python
def is_palindrome(text):
    text = text.lower().replace(' ', '')
    x=''
    for i in range(1,len(text)+1):
        x += text[-i]
    if x==text:
        return True
    else:
        return False
```

Test function on palindrome and non-palindrome.

```python
text = 'A nut for a jar of tuna'
is_palindrome(text)
```

```python
text = 'Not a palindrome'
is_palindrome(text)
```

Now make a recursive version of the function and test it out..

```python
def is_palindrome_rec(text):
    text = text.lower().replace(' ', '')
    if len(text)<2:
        return True
    elif text[0]!=text[-1]:
        return False
    else:
        return is_palindrome_rec(text[1:-1])
```

```python
text = 'A nut for a jar of tuna'
is_palindrome_rec(text)
```

```python
text = 'Not a palindrome'
is_palindrome_rec(text)
```

Create a giant string of letters and compare the time to run both functions.

```python
import time
import random

letters = 'ab'
text = ''.join([random.choice(letters) for x in range(1000000)])
text
```

```python
now = time.time()
print('Without recursion')
print(is_palindrome(text))
print(time.time()-now)

now = time.time()
print('')
print('With recursion')
print(is_palindrome_rec(text))
print(time.time()-now)
```

What about a more efficient version without recursion?

```python
def is_palindrome_better(text):
    text = text.lower().replace(' ', '')
    return (text == text[::-1])
```

```python
now = time.time()
print(is_palindrome_better(text))
print(time.time()-now)
```

Write a function to check if a number is prime with and without recursion. Then compare their performance with a high prime number.

```python
def is_prime(n):
    if n<2:
        return False
    for i in range(2, n//2):
        if (n%i)==0:
            return False
    return True
```

```python
def is_prime_rec(n, i=2): 
    if n==2 or n//2==i:
        return True
    elif n<2 or n%i==0:
        return False
    else:
        return is_prime_rec(n, i+1)
```

```python
import time
number = 1

now = time.time()
print(is_prime(number))
print(time.time()-now)

now = time.time()
print(is_prime_rec(number))
print(time.time()-now)
```

Write a function to flatten a list.

```python
simple_list = [2*[2*[random.choice(range(10)) for x in range(3)] for y in range(3)] for z in range(3)]
simple_list
```

```python
def flatten_list(x):
    flattened = []
    for i in x:
        for j in i:
            for k in j:
                flattened.append(k)
    return flattened
```

```python
flatten_list(simple_list)
```

```python
complex_list = [random.choice([1,2,3])*[random.choice([1,2,3])*[random.choice([1,2,3])*[random.choice(range(10)) for x in range(3)] for y in range(5)] for z in range(5)] for a in range(5)]
complex_list
```

```python
flatten_list(complex_list)
```

```python
def flatten_list_rec(x):
    flat_list=[]
    for i in x:
        if type(i)==list:
            flat_list += (flatten_list_rec(i))
        else:
            flat_list.append(i)
    return flat_list
```

```python
flatten_list_rec(complex_list)
```

#### Using recursion, solve the following problem:

<img src='horses.jpg'>

You are looking to buy some horses. Down the old town road, there is a special deal going on at the stable. Every time you buy a horse, you get a ticket (even if it the horse was free). You can then exchange 7 tickets for an additional horse, free of charge. If you start with \\$1,000,000 and 15 tickets, and horses cost \\$1,150 each, how many total horses can you get?

<b>Hint</b>: You can import the math library and use math.floor to round down a division to the nearest whole number.

```python
import math
```

```python
price = 1150
tickets_per_horse = 7
def buy_horses(money,tickets):
    horses_from_tickets = math.floor(tickets/tickets_per_horse)
    
    horses_from_money = math.floor(money/price)
    money = money - horses_from_money*price
    
    horses_bought = horses_from_tickets + horses_from_money
    
    tickets_earned = horses_bought
    tickets = tickets - horses_from_tickets*tickets_per_horse + tickets_earned

    return horses_bought, money, tickets

new_horses, money, tickets = buy_horses(1000000,0)
total_horses = new_horses
while new_horses>0:
    new_horses, money, tickets = buy_horses(money, tickets)
    total_horses += new_horses
print(total_horses)
```

```python
price = 1150
tickets_per_horse = 7
def buy_horses_rec(money,tickets):
    if tickets < tickets_per_horse and money < price:
        return 0
    else:
        horses_from_tickets = math.floor(tickets/tickets_per_horse)

        horses_from_money = math.floor(money/price)
        money = money - horses_from_money*price

        horses_bought = horses_from_tickets + horses_from_money

        tickets_earned = horses_bought
        tickets = tickets - horses_from_tickets*tickets_per_horse + tickets_earned

        return horses_bought + buy_horses_rec(money, tickets)
```

```python
buy_horses_rec(1000000,15)
```

## <center>Bonus / Homework

Using recursion, explore all elements of 'old_town_road.json', creating a DataFrame of all sub-elements and their data type. Also include their parent element. <br> <br>
Note, you will need to explore down both lists and dictionaries.

```python
df_data = []
def store_data(parent, data_type, name):
    if [parent, data_type, name] not in df_data:
        df_data.append([parent, data_type, name])

def recur(data, parent=None):
    data_type_parent = type(data)
    if data_type_parent == dict:
        for i in data.keys():
            data_type_child = type(data[i])
            store_data(parent, data_type_child, i)
            if data_type_child == dict or data_type_child == list:
                recur(data[i], i)
                
    elif data_type_parent == list:
        for i in data:
            data_type_child = type(i)
            if data_type_child == dict or data_type_child == list:
                recur(i, parent)
```

```python
recur(data)
```

```python
df = pd.DataFrame(df_data)
df.columns = ['Parent', 'Data Type', 'Name']
df
```
